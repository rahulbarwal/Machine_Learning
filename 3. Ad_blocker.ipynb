## Prep Data

#### Remove empty values / nan values

```python
print df.isnull().values.any()

def to_num(cell):
    try: 
        return np.float(cell)
    except: 
        return np.nan

# apply to_num method to every panda series
def series_to_num(series):
    return series.apply(to_num)

train_data = df.iloc[0:, : -1].apply(series_to_num)
```

```python
print train_data.head(10)
```

```python
train_data = train_data.dropna()
```

```python
train_data.head(10)
```

### Convert ad. to 1 and all else to 0

```python
def to_label(s):
    if s == 'ad.':
        return 1
    return 0

train_labels = df.iloc[train_data.index, -1].apply(to_label)
```

## Support Vector machines Algorithm

```python
from sklearn.svm import LinearSVC

svc_model = LinearSVC()
svc_model.fit(train_data[100:2300], train_labels[100:2300])
```

### Testing on random rows

```python
svc_model.predict(train_data.iloc[12].values.reshape(1,-1))
```
